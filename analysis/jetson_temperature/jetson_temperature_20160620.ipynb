```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('companion-20160620-combined.log', delim_whitespace=True )
data['relatime'] = data['system_clock'] - data['system_clock'][0]
print(data.dtypes)
```

```python
data.tail(50)
```

```python
fig = plt.figure( figsize=(16,8), dpi=100 )
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

data.plot( ax=axes, x='relatime', y='bmp280_temp',style='.')
data.plot( ax=axes, x='relatime', y='GPU-therm',style='.')
data.plot( ax=axes, x='relatime', y='CPU-therm',style='.')
ax.set_ylabel('Temp (deg C)')
ax.set_xlabel('System_clock (deg)')
```
